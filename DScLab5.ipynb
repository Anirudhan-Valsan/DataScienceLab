```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
# Load the dataset
train_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DScML LAB PROGRAMS /Sales_Train - Sales_Train.csv")
test_data = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/DScML LAB PROGRAMS /Sales_Test - Sales_Test.csv")
```

```python
train_data.head()
```

```python
train_data.describe()
```

```python
train_data.info()
```

```python
# checking whether there is missing values

train_data.isnull().sum()
```

```python
train_data.fillna(0, inplace=True)
test_data.fillna(0, inplace=True)
```

```python
categorical_cols = ['Item_Identifier', 'Item_Fat_Content', 'Item_Type', 'Outlet_Identifier', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type']
for col in categorical_cols:
    train_data[col] = train_data[col].astype(str)
    test_data[col] = test_data[col].astype(str)
```

Ensure that categorical columns are of string data type

```python
le = LabelEncoder()
for col in categorical_cols:
    train_data[col] = le.fit_transform(train_data[col])
    test_data[col] = le.transform(test_data[col])
```

Encode categorical features

```python
X = train_data.drop('Item_Outlet_Sales', axis=1)
y = train_data['Item_Outlet_Sales']
```

```python
train_data.head()
```

Split the data

```python
k_value = 5
```

choosing k value

```python
knn = KNeighborsRegressor(n_neighbors=k_value)
knn.fit(X, y)
```

Training of a k-NN model

```python
test_predictions = knn.predict(test_data)
```

Make predictions for the test data

```python
submission = pd.DataFrame({'Item identifier': test_data['Item_Identifier'],'Outlet_Identifier': test_data['Outlet_Identifier'],'Item_Outlet_sales': test_predictions})
```

Create a submission file

```python
submission.to_csv('sales_predictions.csv', index=False)
```
