```python
p=sdf['YearsExperience']
q=sdf['Salary']
plt.plot(p,q,label="Experience-Salary")
plt.plot()

plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.title("YE-Salary")
plt.legend()
plt.show()
```

```python
#splitting the data set into training and test
X=sdf.iloc[:,:-1].values
y=sdf.iloc[:,1].values

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
#creating a linear regression model
model=LinearRegression()
```

```python
#fitting the model
model.fit(X_train,y_train)
```

```python
#making prediction
y_pred=model.predict(X_test)
```

```python
#calculating the mean square error
mse=np.mean((y_test-y_pred)**2)
```

```python
print("The mean squared error is ",mse)
```

```python
from sklearn.metrics import r2_score

r2_train=r2_score(y_train,model.predict(X_train))
r2_test=r2_score(y_test,y_pred)


print("Train score",r2_train)
print("Test score",r2_test)
```

```python
# Plot the actual and predicted sales
plt.scatter(X_train, y_train, label='Actual salary')
plt.plot(X_train,model.predict(X_train),color="red", label='Predicted salary')
plt.legend()
plt.xlabel('YearExperience')
plt.ylabel('Salary')
plt.title('Actual vs. predicted salary')
plt.show()
```
