```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_accuracy
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/iris.csv')
```

```python
df.head()
```

```python
X=df.drop('variety',axis=1)
y=df['variety']
```

```python
y=pd.get_dummies(y)
```

```python
scaler=StandardScaler()

X=scaler.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.30,random_state=42)
```

```python
# Define ANN model architecture
model = Sequential()
model.add(Dense(10, input_shape=(4,), activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
# Compile the model
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer, loss='categorical_crossentropy', metrics=[categorical_accuracy])
```

```python
# Train the model
model.fit(X_train, y_train, batch_size=16, epochs=100)
```

```python
# Evaluate the model on testing data
loss, accuracy = model.evaluate(X_test, y_test)
print('Test loss:', loss)
print('Test accuracy:', accuracy)
```

```python
# Make predictions on new data
new_data = np.array([[5.1, 3.5, 1.4, 0.2]])
new_data = scaler.transform(new_data)
prediction = model.predict(new_data)
predicted_class = np.argmax(prediction[0])
print('Predicted class:', df['variety'].iloc[predicted_class])
```
