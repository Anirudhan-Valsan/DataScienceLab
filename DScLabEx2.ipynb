```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("/content/iris.csv")

df.head()
```

```python
#Histogram

plt.figure(figsize=(8,6))
sns.histplot(df['sepal.length'],bins=20,kde=True)
plt.xlabel("sepal.length(cm)")
plt.ylabel("frequency")
plt.title("Histogram of sepal length")
plt.show()
```

```python
#Quartile (Box) plot

plt.figure(figsize=(6,4))
sns.boxplot(x="variety",y='sepal.length',data=df)
plt.xlabel("variety")
plt.ylabel("sepal.length(cm)")
plt.title("Box plot of sepal length by variety")
plt.show()
```

```python
#Distribution chart

plt.figure(figsize=(6,4))
sns.kdeplot(df["sepal.width"],fill=True)
plt.xlabel("sepal.width(cm)")
plt.ylabel("Density")
plt.title("Distribution chart of sepal width")
plt.show()
```

```python
#Scatter plot

plt.figure(figsize=(6,4))
sns.scatterplot(x="sepal.length",y="sepal.width", hue="variety",data=df)
plt.xlabel("sepal length(cm)")
plt.ylabel("sepal width(cm)")
plt.title("Scatter plot of sepal length vs sepal width")
plt.show()
```

```python
#Scatter plot multiple

plt.figure(figsize=(6,4))
sns.scatterplot(x="petal.length",y="petal.width",hue="variety",size="sepal.length",data=df)
plt.xlabel("Petal length(cm)")
plt.ylabel("Petal width(cm)")
plt.title("Scatter plot of petal length vs petal width colored by variety,sized by sepal length")
plt.show()
```

```python
#scatter matrix(pair plot)

sns.set(style="ticks")
sns.pairplot(df,hue="variety",markers=["o","s","d"])
plt.suptitle("scatter matrix(pair plot)",y=1.02)
plt.show()
```

```python
#Bubble chart

plt.figure(figsize=(6,4))
sns.scatterplot(x="sepal.length",y="sepal.width",hue="variety",size="petal.length",data=df)
plt.xlabel("sepal length(cm)")
plt.ylabel("sepal width(cm)")
plt.title("Bubble chart of sepal length vs sepal width(colored by variety,sized by petal length)")
plt.show()
```

```python
#Density chart

plt.figure(figsize=(6,4))
sns.kdeplot(x=df["petal.length"],y=df["petal.width"],cmap="Blues",fill=True)
plt.xlabel("petal length(cm)")
plt.ylabel("petal width(cm)")
plt.title("Density chart of petal length vs petal width")
plt.show()
```

```python
#Parallel coordinates chart

plt.figure(figsize=(6,4))
sns.set(style="whitegrid")
pd.plotting.parallel_coordinates(df,"variety",colormap="viridis")
plt.title("parallel coordinates chart")
plt.show()
```

```python
#Deviation chart

feature="sepal.length"
mean_value=df[feature].mean()
df["deviation"]=df[feature]-mean_value
plt.figure(figsize=(6,4))
plt.plot(df.index,df["deviation"],marker='o',linestyle='-',color='b')
plt.xlabel("Data points")
plt.ylabel("deviation from mean{features.capitalize()}(cm)")
plt.title(f"deviation chart for {feature.capitalize()}")
plt.grid(True)
plt.show()
```

```python
#Andrews curves

plt.figure(figsize=(6,4))
sns.set(style="ticks")
pd.plotting.andrews_curves(df,"variety",colormap="viridis")
plt.title("Andrews curves")
plt.show()
```
